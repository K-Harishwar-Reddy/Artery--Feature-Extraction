```python
import pandas as pd
import numpy as np

# Function to load and format the sheet
def load_and_format_sheet(sheet_name, file_path, skip_rows=1):
    df = pd.read_excel(file_path, sheet_name=sheet_name, skiprows=skip_rows)
    df = df.rename(columns={'Unnamed: 0': 'Artery_ID'}).set_index("Artery_ID")
    return df
```

```python
# From Jeff
PATH_LABEL = "/Users/jinzhou/Desktop/Paper_MIA/data/Batch 1,2,3 Art Scores COMPLETE.xlsx"

# Load each batch
df_label_batch_1 = load_and_format_sheet(sheet_name=0, file_path=PATH_LABEL)
df_label_batch_2 = load_and_format_sheet(sheet_name=1, file_path=PATH_LABEL)
df_label_batch_3 = load_and_format_sheet(sheet_name=2, file_path=PATH_LABEL)

# Concatenate batches and replace values for Jeff's data
df_label_jeff = pd.concat([df_label_batch_1, df_label_batch_2, df_label_batch_3], axis=1)
jeff_replacement_dict = {
    "without arteriosclerosis": 0, 
    "mild arteriosclerosis": 1,
    "mild hyalinosis": np.nan, 
    "moderate arteriosclerosis": 2,
    "severe arteriosclerosis": 3,
    "0 (no green line - reject)": 0, 
    "1 (no green line - reject)": 1,
    "reject, no lumen": np.nan, 
    "fix": np.nan, 
    "-": np.nan, 
    " - ": np.nan
}
df_label_jeff = df_label_jeff.replace(jeff_replacement_dict)
df_label_jeff.to_csv("/Users/jinzhou/Desktop/Paper_MIA/data/labels_jeff.csv")
```
